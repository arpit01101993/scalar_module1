```python
product_rating = pd.merge(products, orders, on='product_id', how='left')
product_rating
```

```python
product_rating['review_title'].head()
```

```python
def uppercase_title(title):
  return title.upper()
```

```python
product_rating['title_upper'] = product_rating['review_title'].apply(uppercase_title)
```

```python
product_rating['title_upper'].head()
```

```python
product_rating.info()
```

```python
def extract_price(price_str):
  try:
    return float(str(price_str).replace('₹','').replace(',',''))
  except (ValueError, TypeError):
    return None
```

```python
product_rating['actual_price'] = product_rating['actual_price'].apply(extract_price)
product_rating['discounted_price'] = product_rating['discounted_price'].apply(extract_price)
```

```python
print(product_rating['actual_price'].head())
print(product_rating['discounted_price'].head())
```

```python
def calculate_discount_percentage(row):
  actual_price = row['actual_price']
  discounted_price = row['discounted_price']
  return ((actual_price - discounted_price) / actual_price) * 100 if actual_price else None

product_rating['calculated_discount'] = product_rating.apply(calculate_discount_percentage, axis=1)
```

```python
product_rating['calculated_discount'].head()
```

```python
def max_rating(series):
  return series.max()

def min_rating_count(series):
  return series.min()
```

```python
rating_stats = product_rating[['rating','rating_count']].apply([max_rating, min_rating_count])
rating_stats
```

```python
def avg_price_diff(group):
  avg_actual = group['actual_price'].mean()
  avg_discounted = group['discounted_price'].mean()
  return avg_actual - avg_discounted

price_diff_by_category = product_rating.groupby('category').apply(avg_price_diff)
price_diff_by_category
```

```python
def avg_percentage_discount(group):
  avg_actual = group['actual_price'].mean()
  avg_discounted = group['discounted_price'].mean()
  return ((avg_actual - avg_discounted)/avg_actual)*100 if avg_actual else None

percentage_discount_by_category = product_rating.groupby('category').apply(avg_percentage_discount)
percentage_discount_by_category
```

```python
import matplotlib.pyplot as plt

percentage_discount_by_category.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.title('Average Discount Percentage by Category')
plt.show()
```
